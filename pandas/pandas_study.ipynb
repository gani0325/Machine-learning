## lambda, map, apply

### lambda 함수

##### 한 줄로 함수를 표현하는 익명함수 기법
###### lambda argument : expression
###### lambda x, y : x+y
###### def f(x,y) :
###### return x+y

###### f(1,4)

```python
def f(x,y) :
    return x * y
f(4,5)
```

```python
f = lambda x: x/2
f(3)
```

```python
(lambda x : x+2)(6)
```

### map 함수 (list 꼭 쓰기)
##### 함수와 sequence형 데이터를 인자로 받아
##### 각 element 마다 입력받은 함수를 적용하여 list로 반환(주로 함수 lambda)
##### map(funcion : sequence)

```python
ex = [2,3,5,6,7]
```

```python
f = lambda x, y : x+y
list(map(f,ex,ex))
```

```python
list(map(lambda x : x+10 , ex ))
```

```python
s11 = Series(np.arange(10))
s11
```

```python
s11.map(lambda x : x **2)
```

```python
z = {1:'a', 2: 'b', 3:'c'}
s11.map(z)
```

```python
s22 = Series(np.arange(10,20))
s11.map(s22)
```

### apply for dataframe
##### map 과 달리, series 전체(column)에 해당 함수 적용
##### 입력값이 series 데이터로 입력받아 handling 가능
##### 각 column 별로 계산
##### 내장함수 가능
##### scalar 값 이외의 series 값의 반환 가능

```python
df_info = df[['earn','height']]
df_info.head()
```

```python
f = lambda x : x.max() - x.min()
df_info.apply(f)
```

```python
df_info.sum()
```

```python
df_info.apply(sum)
```

```python
def f(x) :
    return Series([x.min(), x.max()], index = ['min', 'max'])
df_info.apply(f)
```

### applymap for dataframe
##### series 가 아닌 element 단위로 함수 적용
##### apply 적용과 같음

```python
f = lambda x: -x
df_info.applymap(f).head()
```

```python
f = lambda x : -x
df_info[['earn','height']].apply(f).head()
```

#### unipue
##### series data의 유일한 값 list 반환

```python
df.race.unique()
```

```python
np.array(dict(enumerate(df['race'].unique()))) # dict type으로 index
```

```python
value = list(map(int, np.array(list(enumerate(df["race"].unique())))[:, 0].tolist()))
key = np.array(list(enumerate(df["race"].unique())), dtype=str)[:, 1].tolist()

value, key
```

```python
df["race"].replace(to_replace=key, value=value, inplace=True)
```

```python
value = list(map(int, np.array(list(enumerate(df["sex"].unique())))[:, 0].tolist()))
key = np.array(list(enumerate(df["sex"].unique())), dtype=str)[:, 1].tolist()

value, key
```

```python
df["sex"].replace(to_replace=key, value=value, inplace=True)
df.head(5)
```
